```python
import wrds
conn = wrds.Connection()
```

```python
conn.list_tables(library='ciq')
```

```python
data = conn.get_table(library='ciq', table='ciqtranscript')
```

```python
data = data.reset_index()
```

```python
data.to_feather('data/earningscall/ciq_transcript.feather')
```

```python
data = conn.get_table(library='ciq', table='wrds_transcript_person', obs=10)
```

```python
data
```

```python
import pandas as pd
splist = pd.read_csv('data/earningscall/earningscall_list.csv')

for i in range(splist.shape[0]):
    tranid = splist['transcriptid'][i]
    raw_sql = """select * from ciq.ciqtranscriptcomponent where transcriptid =  """ + str(tranid)
    component = conn.raw_sql(raw_sql, date_cols=['date']).sort_values('componentorder').reset_index(drop=True)
    raw_sql = """select * from ciq.wrds_transcript_person where transcriptid =  """ + str(tranid)
    spkear = conn.raw_sql(raw_sql, date_cols=['date']).sort_values('componentorder').reset_index(drop=True)
    spkear['componenttext'] = component['componenttext']
    spkear.to_feather()
```
